```python
import pandas as pd
import matplotlib.pyplot as mpl
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
dataset=pd.read_csv('website_data_20190225.csv')
```

```python
dataset.head(5)
```

```python
dataset.drop(['STATE','DISTRICT','TEH_NAME','LAT','LON','BLOCK_NAME','WLCODE'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
st=pd.get_dummies(dataset['SITE_TYPE'],drop_first=True)
st.head()
```

```python
sn=pd.get_dummies(dataset['SITE_NAME'],drop_first=True)
sn.head()
```

```python
test=dataset
```

```python
test.drop(['SITE_NAME','SITE_TYPE'],axis=1,inplace=True)
```

```python
test=pd.concat([sn,st,test],axis=1)
```

```python
test.head()
```

```python
abc=test
```

```python
type(abc)
```

```python
abc.replace(to_replace="'0",value=0,inplace=True)
```

```python
type(abc['MONSOON'][708])
```

```python
# for i in range(0,abc.shape[0]):
#     abc['MONSOON'][i]=float(abc['MONSOON'][i])
#     abc['POMRB'][i]=float(abc['POMRB'][i])
#     abc['POMKH'][i]=float(abc['POMKH'][i])
#     abc['PREMON'][i]=float(abc['PREMON'][i])
```

```python
abc['MONSOON'][708]
```

```python
first=list(abc['MONSOON'])
second=list(abc['POMRB'])
third=list(abc['POMKH'])
fourth=list(abc['PREMON'])
s=[x+y+z+w for x, y,z,w in zip(first, second, third, fourth)]
pd.core.frame.DataFrame(s)
s[708]
```

```python
abc['MONSOON']=s
```

```python
abc.drop(['POMRB','POMKH','PREMON'],axis=1,inplace=True)
```

```python
abc.head()
```

```python
x=abc.drop('MONSOON',axis=1)
y=abc['MONSOON']
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
from sklearn.linear_model import LinearRegression
ob=LinearRegression()
ob.fit(x_train,y_train)
pred=ob.predict(x_test)
co=pred
j=0
for i in co:
    if i<0:
        co[j]=0
    j+=1
df = pd.DataFrame({'Actual': y_test, 'Predicted': co})  
df.head()
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, co))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, co))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, co)))
```
