# Testing Stationarity

```python
def adf_test(series,title=''):
    
    print(f'Augmented Dickey-Fuller Test: {title}')
    result = adfuller(series.dropna(),autolag='AIC') 
    
    labels = ['ADF test statistic','p-value','# Lags used','# Observations']
    out = pd.Series(result[0:4],index=labels)

    for key,val in result[4].items():
        out[f'Critical Value ({key})']=val
        
    print(out.to_string())          
    
    if result[1] <= 0.05: print("Data has no unit root and is stationary")
    else: print("Data has a unit root and is non-stationary")
```

```python
def test_stationary(timeseries):
    
    moving_average=timeseries.rolling(window=12).mean()
    standard_deviation=timeseries.rolling(window=12).std()
    
    plt.plot(timeseries,color='blue',label="Original")
    plt.plot(moving_average,color='red',label='Mean')
    plt.plot(standard_deviation,color='black',label='Standard Deviation')
    plt.legend(loc='best')  #best for axes
    plt.title('Rolling Mean & Deviation')
    plt.show(block=False)
    
    print('Results Of Dickey-Fuller Test')
    tstest=adfuller(timeseries.iloc[:,0],autolag='AIC')
    tsoutput=pd.Series(tstest[0:4],index=['Test Statistcs','P-value','#Lags used',"#Obs. used"])
    
    for key,value in tstest[4].items():
        tsoutput['Critical Value (%s)'%key]=value
    
    print((tsoutput))
```

## Preprocessed Data

```python
indexedDataset = df.copy()
```

```python
indexedDataset.head()
```

```python
plt.plot(indexedDataset, label='Water level over past years');
plt.xlabel('Years');
plt.ylabel('Water-Level');
plt.title('Total rainfall over the past years');
plt.legend();
plt.show();
```

```python
test_stationary(indexedDataset)
```

```python
adf_test(indexedDataset)
```

## Indexed Dataset Logscale

```python
indexedDataset_logscale = np.log(indexedDataset)
```

```python
test_stationary(indexedDataset_logscale)
```

```python
adf_test(indexedDataset_logscale)
```

## Log Minus Moving Average (dl_ma)

```python
rolmeanlog = indexedDataset_logscale.rolling(window=12).mean()
```

```python
dl_ma = indexedDataset_logscale - rolmeanlog
dl_ma.head(12)
```

```python
dl_ma.dropna(inplace=True)
dl_ma.head(12)
```

```python
test_stationary(dl_ma)
```

```python
adf_test(dl_ma)
```

## Exponential Decay Weighted Average (edwa)

```python
edwa = indexedDataset_logscale.ewm(halflife=12,min_periods=0,adjust=True).mean()
```

```python
plt.plot(indexedDataset_logscale, label='Log scaled');
plt.plot(edwa, color='red', label='Exponential');
plt.xlabel('Years');
plt.ylabel('Water-Level');
plt.title('Total rainfall over the past years');
plt.legend();
plt.show();
```

## Logscale - Moving Exponential Decay Average (dlmeda)

```python
dlmeda = indexedDataset_logscale - edwa
```

```python
test_stationary(dlmeda)
```

```python
adf_test(dlmeda)
```

## Log Div Shifting (dlds)

```python
# Before Shifting
indexedDataset_logscale.head()
```

```python
# After Shifting

indexedDataset_logscale.shift().head()
```

```python
dlds = indexedDataset_logscale - indexedDataset_logscale.shift()
```

```python
dlds.dropna(inplace=True)

test_stationary(dlds)
```

```python
adf_test(dlds)
```

## Decomposition

```python
decompostion= seasonal_decompose(indexedDataset_logscale,freq=10)
```

```python
trend = decompostion.trend
seasonal = decompostion.seasonal
residual = decompostion.resid
```

```python
decompostion.plot()
```

```python
plt.subplot(411)
plt.plot(indexedDataset_logscale,label='Original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(trend,label='Trend')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(seasonal,label='Seasonal')
plt.legend(loc='best')

plt.subplot(414)
plt.plot(residual,label='Residual')
plt.legend(loc='best')

plt.tight_layout() 
```

```python
decomposedlogdata = residual
```

```python
decomposedlogdata = pd.DataFrame(decomposedlogdata.dropna())
test_stationary(decomposedlogdata)
```

```python
adf_test(decomposedlogdata)
```

```python
decomposedlogdata
```

# Modeling

## ACF & PACF Plots

```python
lag_acf = acf(dlds,nlags=20)
lag_pacf = pacf(dlds,nlags=20,method='ols')
```

```python
plot_acf(dlds);
plt.show()
```

```python
plot_pacf(dlds);
plt.show()
```

```python
plt.subplot(121);
plt.plot(lag_acf);
plt.axhline(y=0, linestyle='--',color='gray');
plt.axhline(y=1.96/np.sqrt(len(dlds)),linestyle='--',color='gray');
plt.axhline(y=-1.96/np.sqrt(len(dlds)),linestyle='--',color='gray');
plt.title('AutoCorrelation Function');

plt.subplot(122);
plt.plot(lag_pacf);
plt.axhline(y=0, linestyle='--',color='gray');
plt.axhline(y=1.96/np.sqrt(len(dlds)),linestyle='--',color='gray');
plt.axhline(y=-1.96/np.sqrt(len(dlds)),linestyle='--',color='gray');
plt.title('PartialAutoCorrelation Function');

plt.tight_layout();
```

## ARIMA(5,1,0) ~ AR Modeling

```python
model = ARIMA(indexedDataset_logscale,order=(5,1,0))
```

```python
results_AR = model.fit(disp=-1)
```

```python
plt.plot(dlds, label='DLDS');
plt.plot(results_AR.fittedvalues,color='red', label='AR-Results');
plt.legend();
plt.title('Residual Sum of Squares: %.4f'%sum((results_AR.fittedvalues-dlds['Total'])**2));
plt.show();
```

```python
results_AR.summary()
```

## ARIMA(0,1,2) ~ MA Modeling

```python
model = ARIMA(indexedDataset_logscale, order=(0, 1, 2))
```

```python
results_MA = model.fit(disp=-1)  
```

```python
plt.plot(dlds, label='DLDS');
plt.plot(results_MA.fittedvalues, color='red', label='MA-Results');
plt.title('RSS: %.4f'%sum((results_MA.fittedvalues-dlds['Total'])**2));
plt.legend();
plt.show();
```

## ARIMA(5,1,2)

```python
model = ARIMA(indexedDataset_logscale, order=(5, 1, 2))  
```

```python
results_ARIMA = model.fit(disp=-1)
```

```python
plt.plot(dlds, label='DLDS');
plt.plot(results_ARIMA.fittedvalues, color='red', label='ARIMA-Results');
plt.title('RSS: %.4f'%sum((results_ARIMA.fittedvalues-dlds['Total'])**2));
plt.legend();
plt.show();
```

# Predictions

## Making Predictions for entire df

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff.head()
```

## Rescaling to original scale

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_diff_cumsum.head()
```

```python
predictions_ARIMA_log = pd.Series(indexedDataset_logscale['Total'], index=indexedDataset_logscale.index)
predictions_ARIMA_log
```

```python
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
```

## Visualization

```python
plt.plot(indexedDataset, label='Original');
plt.plot(predictions_ARIMA, color='red',label='ARIMA(5,1,2)')
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-indexedDataset['Total'])**2)/len(indexedDataset)))
plt.legend();
plt.show();
```
